# step 3: normalization of data

```python
X=df["LSTAT"].values
Y=df["target"].values

#creating an instance object then using fit tranform to 
x_scaler=MinMaxScaler()
X=x_scaler.fit_transform(X.reshape(-1,1))
X=X[:,-1]

y_scaler=MinMaxScaler()
Y=y_scaler.fit_transform(Y.reshape(-1,1))
Y=Y[:, -1]
```

# Step 4: Spliting the data

```python
#spliting the data
#0.2 indicated spliting 20% of the dataset into dataset
xtrain ,xtest, ytrain, ytest= train_test_split(X,Y, test_size=0.2)
xtrain=xtrain.reshape(-1,1)
ytrain=ytrain.reshape(-1,1)
xtest=xtest.reshape(-1,1)
ytest=ytest.reshape(-1,1)
```

# Mean sqared error

```python
def error (m,x,c,t):
    N=x.size
    e=sum(((m * x + c)-t) ** 2)
    return e * 1 / (2 * N)
```

```python
def update(m,x,c,t,learning_rate):
    grad_m=sum(2*((m*x + c)-t) *x)
    grad_c=sum(2*((m*x + c)-t))
    m=m-grad_m*learning_rate
    c=c-grad_c*learning_rate
    return m,c
    
    
```

```python
def gradient_descent(init_m,init_c,x,t,learning_rate,iterations,error_threshold):
    m=init_m
    c=init_c
    error_values=list()
    mc_values=list()
    for i in range(iterations):
        e=error(m,x,c,t)
        if e<error_threshold:
            print("error is less than threshold. stopping gradient descent")
            break
        error_values.append(e)
        m,c=update(m,x,c,t,learning_rate)
        mc_values.append((m,c))
    return m,c,error_values,mc_values
```

# using gradient descent

```python
%%time
init_m=0.9
init_c=0
learning_rate=0.001
iterations=250
error_threshold=0.001
m,c,error_values,mc_values=gradient_descent(init_m,init_c,xtrain,ytrain,learning_rate,iterations,error_threshold)
```

```python
#plot the regrassion curve
plt.scatter(xtrain,ytrain,color="b")
plt.plot(xtrain,(m*xtrain+c),color="r")
```

```python
#plot error
plt.plot(np.arange(len(mc_values)),error_values)
plt.ylabel("error")
plt.xlabel("iterations")
```
